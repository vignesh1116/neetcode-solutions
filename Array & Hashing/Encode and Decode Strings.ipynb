# 008. Encode and Decode Strings
🔗 https://neetcode.io/problems/string-encode-and-decode

## Problem Statement
Design an algorithm to encode a list of strings to a string. The encoded string is then sent over the network and is decoded back to the original list of strings.

## Example

Input: dummy_input = ["Hello","World"]

Output: ["Hello","World"]

Explanation:
Machine 1:
Codec encoder = new Codec();
String msg = encoder.encode(strs);
Machine 1 ---msg---> Machine 2

Machine 2:
Codec decoder = new Codec();
String[] strs = decoder.decode(msg);

## Solution

```python
class Solution:

    def encode(self, strs: List[str]) -> str:
        if not strs:
            return ""
        sizes , result = [], ""
        for word in strs:
            sizes.append(len(word))
        for lengths in sizes:
            result += str(lengths)
            result += "," 
        result += "#"  
        for words in strs:
            result += words
        return result 

    def decode(self, s: str) -> List[str]:
        if not s:
            return []
        sizes, result, i = [], [], 0
        while s[i] != "#" : 
            hashcut = "" 
            while s[i] != ",":
                hashcut += s[i] 
                i += 1
            sizes.append(int(hashcut)) 
            i += 1 
        i += 1 
        for size in sizes: 
            result.append(s[i:i+size])
            i += size 
        return result
```

Store string length + delimiter

Read length
Extract string

Key idea: Avoid ambiguity using length prefix
